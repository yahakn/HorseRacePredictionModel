```python
import re

def temizle_yas(x):
    if pd.isna(x):
        return x  # NaN ise dokunma
    x = str(x).strip()  # BaÅŸ ve son boÅŸluklarÄ± temizle
    match = re.search(r'\d+', x)  # Ä°lk geÃ§en rakamÄ± bul
    if match:
        return int(match.group())  # Bulduysa integer yap
    else:
        return None  # SayÄ± bulamazsa None yap

# ðŸ“Œ UygulayalÄ±m
df["YaÅŸ"] = df["YaÅŸ"].apply(temizle_yas)
```

```python
import pandas as pd
data = df
# ðŸ“Œ DataFrame oluÅŸturuyoruz
df = pd.DataFrame(data)

# ðŸ“Œ At ismiyle ilgili olmayan kÄ±saltmalar listesi
kisaltmalar = ["KG", "KUL", "DB", "BB", "Ã‡ABA", "GKR", "SKG", "SKUL", "SGKR", "YP", "TGK", "Ã–G", "DS", "SK", "K"]

# ðŸ“Œ At ismi ve detaylarÄ± ayÄ±ran fonksiyon (while ile)
def ayir_at_isim(at_isim):
    kelimeler = at_isim.split()
    detaylar = []

    # Sondan baÅŸlayarak, kÄ±saltmalarÄ± ayÄ±r
    while kelimeler and kelimeler[-1] in kisaltmalar:
        detaylar.append(kelimeler.pop())  # Son kelimeyi detaylara ekle

    return " ".join(kelimeler), " ".join(reversed(detaylar)) if detaylar else None

# ðŸ“Œ Yeni sÃ¼tunlarÄ± ekleyelim
df[["At Ä°smi", "Detaylar"]] = df["At Ä°smi"].apply(lambda x: pd.Series(ayir_at_isim(x)))

# ðŸ“Œ SonuÃ§larÄ± gÃ¶relim
df
```

```python
def kÄ±salt(jokey_adi):
    isim_soyisim = jokey_adi.split()  # Ä°sim ve soyadÄ± ayÄ±rÄ±yoruz
    kÄ±saltma = None  # kÄ±saltma deÄŸiÅŸkenini baÅŸtan tanÄ±mlÄ±yoruz

    if len(isim_soyisim) == 3:  # EÄŸer 2 isim ve 1 soyadÄ± varsa (Ã–rnek: Yusuf Hakan AktaÅŸ)
        kÄ±saltma = isim_soyisim[0][0].upper() + '.' + isim_soyisim[1][0].upper() + '.' + isim_soyisim[2].upper()  # Ä°ki ismin baÅŸ harfini alÄ±p soyadÄ± yazÄ±yoruz
    elif len(isim_soyisim) == 2:  # EÄŸer 1 isim ve 1 soyadÄ± varsa (Ã–rnek: Mehmet Kaya)
        kÄ±saltma = isim_soyisim[0][0].upper() + '.' + isim_soyisim[1].upper()  # AdÄ± ve soyadÄ±nÄ± kÄ±saltÄ±yoruz
    else:  # EÄŸer beklenmedik bir durum varsa (Ã¶rneÄŸin yalnÄ±zca bir kelime varsa)
        kÄ±saltma = jokey_adi.upper()  # Hala bir ÅŸey yapÄ±lmadÄ±ysa tam adÄ± dÃ¶ndÃ¼rÃ¼yoruz.

    return kÄ±saltma
```

```python
jokey['Jokey KÄ±saltmasÄ±'] = jokey['Jokey'].apply(kÄ±salt)
```

```python
# 1. Jokey kÄ±saltmalarÄ±nÄ± ve tam adlarÄ±nÄ± iÃ§eren bir sÃ¶zlÃ¼k oluÅŸturuyoruz
jokey_dict = dict(zip(jokey['Jokey KÄ±saltmasÄ±'], jokey['Jokey']))

# 2. YarÄ±ÅŸlar tablosundaki 'Jokey AdÄ±' kÄ±saltmalarÄ±nÄ± bu sÃ¶zlÃ¼kle tam adÄ±yla deÄŸiÅŸtiriyoruz
df['Jokey'] = df['Jokey AdÄ±'].map(jokey_dict)

# 3. Sonucu kontrol edelim
print(df[['Jokey AdÄ±', 'Jokey']].head())
```
