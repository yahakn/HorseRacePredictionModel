```python
import re

def temizle_yas(x):
    if pd.isna(x):
        return x  # NaN ise dokunma
    x = str(x).strip()  # Baş ve son boşlukları temizle
    match = re.search(r'\d+', x)  # İlk geçen rakamı bul
    if match:
        return int(match.group())  # Bulduysa integer yap
    else:
        return None  # Sayı bulamazsa None yap

# 📌 Uygulayalım
df["Yaş"] = df["Yaş"].apply(temizle_yas)
```

***At İsmi ve Detaylar***

```python
import pandas as pd
data = df
# 📌 DataFrame oluşturuyoruz
df = pd.DataFrame(data)

# 📌 At ismiyle ilgili olmayan kısaltmalar listesi
kisaltmalar = ["KG", "KUL", "DB", "BB", "ÇABA", "GKR", "SKG", "SKUL", "SGKR", "YP", "TGK", "ÖG", "DS", "SK", "K"]

# 📌 At ismi ve detayları ayıran fonksiyon (while ile)
def ayir_at_isim(at_isim):
    kelimeler = at_isim.split()
    detaylar = []

    # Sondan başlayarak, kısaltmaları ayır
    while kelimeler and kelimeler[-1] in kisaltmalar:
        detaylar.append(kelimeler.pop())  # Son kelimeyi detaylara ekle

    return " ".join(kelimeler), " ".join(reversed(detaylar)) if detaylar else None

# 📌 Yeni sütunları ekleyelim
df[["At İsmi", "Detaylar"]] = df["At İsmi"].apply(lambda x: pd.Series(ayir_at_isim(x)))

# 📌 Sonuçları görelim
df
```

```python
df["Jokey Adı"] = df["Jokey Adı"].str.replace(" AP$", "", regex=True)
```

```python
def kısalt(jokey_adi):
    isim_soyisim = jokey_adi.split()  # İsim ve soyadı ayırıyoruz
    kısaltma = None  # kısaltma değişkenini baştan tanımlıyoruz

    if len(isim_soyisim) == 3:  # Eğer 2 isim ve 1 soyadı varsa (Örnek: Yusuf Hakan Aktaş)
        kısaltma = isim_soyisim[0][0].upper() + '.' + isim_soyisim[1][0].upper() + '.' + isim_soyisim[2].upper()  # İki ismin baş harfini alıp soyadı yazıyoruz
    elif len(isim_soyisim) == 2:  # Eğer 1 isim ve 1 soyadı varsa (Örnek: Mehmet Kaya)
        kısaltma = isim_soyisim[0][0].upper() + '.' + isim_soyisim[1].upper()  # Adı ve soyadını kısaltıyoruz
    else:  # Eğer beklenmedik bir durum varsa (örneğin yalnızca bir kelime varsa)
        kısaltma = jokey_adi.upper()  # Hala bir şey yapılmadıysa tam adı döndürüyoruz.

    return kısaltma
```

```python
jokey['Jokey Kısaltması'] = jokey['Jokey'].apply(kısalt)
```

```python
jokey
```

```python
# 1. Jokey kısaltmalarını ve tam adlarını içeren bir sözlük oluşturuyoruz
jokey_dict = dict(zip(jokey['Jokey Kısaltması'], jokey['Jokey']))

# 2. Yarışlar tablosundaki 'Jokey Adı' kısaltmalarını bu sözlükle tam adıyla değiştiriyoruz
df['Jokey'] = df['Jokey Adı'].map(jokey_dict)

# 3. Sonucu kontrol edelim
print(df[['Jokey Adı', 'Jokey']].head())
```

```python
df_ortak = pd.DataFrame(set(zip(tarih_set["Koşu"], tarih_set["Birinci"], tarih_set["Derece"],tarih_set["Tarih"])),
                         columns=["Koşu", "Birinci", "Derece", "Tarih"])

# 📌 Sonucu yazdır
df_ortak
```

```python
df_ortak_ = pd.DataFrame(set(zip(tarih_set["Koşu"], tarih_set["Birinci"], tarih_set["Derece"],tarih_set["Tarih"],tarih_set["Şehir"],tarih_set["Pist"])),
                         columns=["Koşu", "Birinci", "Derece", "Tarih", "Şehir","Pist"])

# 📌 Sonucu yazdır
df_ortak_
```

```python
# At İsmi ve Derece eşleşmesine göre Tarih ve Koşu'yu alalım
df = df.merge(
    df_ortak[['Birinci', 'Derece', 'Tarih', 'Koşu']],
    left_on=['At İsmi', 'Derece'],  # df'deki 'At İsmi' ile eşleştiriyoruz
    right_on=['Birinci', 'Derece'],  # df_ortak'taki 'Birinci' ile
    how='left',
    suffixes=('', '_ortak')
)

# df_ortak'tan gelen Tarih ve Koşu'yu df'ye kopyalayalım
df['Tarih'] = df['Tarih_ortak']
df['Koşu'] = df['Koşu_ortak']

# Artık ekstra sütunları silelim
df.drop(['Birinci', 'Tarih_ortak', 'Koşu_ortak'], axis=1, inplace=True)

print(df.head())
```

```python
# Tarih ve Koşu sütunlarındaki boşlukları yukarıdan aşağı doldur
df['Tarih'] = df['Tarih'].fillna(method='ffill')
df['Koşu'] = df['Koşu'].fillna(method='ffill')
```

```python
# Tarih, Koşu, Derece değerleri ile eşleşen satırlara Pist ve Şehir sütunlarını df veri setine eklemek için merge işlemi yapıyoruz
df = pd.merge(df, df_ortak_[['Tarih', 'Koşu', 'Derece', 'Şehir', 'Pist']],
              on=['Tarih', 'Koşu', 'Derece'], how='left')

# 📌 Sonucu yazdır
df
```

```python
df['Şehir'] = df['Şehir'].fillna(method='ffill')
df['Pist'] = df['Pist'].fillna(method='ffill')
```
